```python
test_img = cv2.imread('./dataset/test/04.bmp')
plt.imshow(cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB))
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (10, 10))
# kernel = cv2.erode(test_img, kernel, iterations=1)
plt.imshow(cv2.cvtColor(cv2.morphologyEx(test_img, cv2.MORPH_OPEN, kernel), cv2.COLOR_BGR2RGB))
```

```python
alpha=1.5
beta=15

test_img=cv2.addWeighted(test_img,alpha,np.zeros(test_img.shape, test_img.dtype),0,beta)

plt.imshow(test_img)
```

```python
# morphology_ex on test_img
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
morphology_ex_img = cv2.morphologyEx(test_img, cv2.MORPH_CLOSE, kernel, iterations=3)
morphology_ex_img = cv2.erode(morphology_ex_img, kernel, iterations=3)

plt.imshow(cv2.cvtColor(morphology_ex_img, cv2.COLOR_BGR2RGB))
```

```python
# convert to HSV
hsv = cv2.cvtColor(test_img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv, cmap='hsv')
```
